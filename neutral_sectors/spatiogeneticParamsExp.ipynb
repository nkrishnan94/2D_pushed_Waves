```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.optimize import curve_fit
```

```python
df = pd.read_excel('RachaSectors.xlsx',sheet_name = '1ul')
```

```python
colors= plt.cm.viridis(np.linspace(0,1,4))
```

```python
dfR1 = pd.concat([df.iloc[1:6,1:5],
          df.iloc[1:6,6:10],
          df.iloc[1:6,11:15]],axis=1, join="inner")

dfS1 = pd.concat([df.iloc[9:14,1:5],
          df.iloc[9:14,6:10],
          df.iloc[9:14,11:15]],axis=1, join="inner")
dfS1 = dfS1.rename(index = {9:1,10:2,11:3,12:4,13:5})

df1 = pd.concat([dfR1,dfS1],axis=1, join = 'inner')
df1.columns = ['Radius'+str(i) for i in range(1,13)] + ['Sectors'+str(i) for i in range(1,13)] 
```

```python
fig,ax = plt.subplots()
plt.errorbar(x=df1.loc[:,'Radius1':'Radius12'].values.mean(axis=1),
             y=df1.loc[:,'Sectors1':'Sectors12'].values.mean(axis=1),
             xerr = df1.loc[:,'Radius1':'Radius12'].values.astype(float).std(axis=1),
             yerr = df1.loc[:,'Sectors1':'Sectors12'].values.astype(float).std(axis=1),
             ecolor='k',
             elinewidth=2,
            lw=.5,marker='o',c=colors[0],markersize=13)

ax.set_xlabel('Radius (mm)',fontsize=16)
ax.set_ylabel('Sectors',fontsize=16)
#ax.set_xticklabels(ax.get_xticklabels(),fontsize=12)
ax.tick_params(axis='x', labelsize=14)
ax.tick_params(axis='y', labelsize=14)
```
