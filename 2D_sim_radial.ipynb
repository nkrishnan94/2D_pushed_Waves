```python
var_het_files = glob.glob("radial_growth/varhet_K10000_*.txt")
```

```python
b0 =[]
b4 = []
for f in var_het_files:
    if "B0_" in f:
        b0.append(f)
    if "B4_" in f:
        b4.append(f)
```

```python
arr0 = np.zeros((len(b0),300,2))
for i,f in enumerate(b0):
    arr0[i] = np.loadtxt(f,delimiter =  ', ')
    
arr1 = np.zeros((len(b4),300,2))
for i,f in enumerate(b4):
    arr1[i] = np.loadtxt(f,delimiter =  ', ')
```

```python
len(arr0[0,:,0])
```

```python
fig, ax = plt.subplots()
ax.plot(arr0[0,:,0],np.mean(arr0[:,:,1],axis=0),lw=4,label= 'B=0',c='navy')
ax.plot(arr0[0,:,0],np.mean(arr1[:,:,1],axis=0),lw =4,label= 'B=4',c='green')

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_linewidth(2)
ax.spines['bottom'].set_linewidth(2)
ax.tick_params(axis='both', which='major', labelsize=14)
ax.set_xlabel('time (generations)',fontsize=16)
ax.set_ylabel(r'$Var_ij(H)$',fontsize=16)
ax.legend()
```

```python
pop_files = glob.glob("radial_growth/pop_K10000_*.txt")
```

```python
b0p =[]
b4p = []
for f in pop_files:
    if "B0_" in f:
        b0p.append(f)
    if "B4_" in f:
        b4p.append(f)
```

```python
arr0p = np.zeros((len(b0p),300,2))
for i,f in enumerate(b0p):
    arr0p[i] = np.loadtxt(f,delimiter =  ', ')
    
arr1p = np.zeros((len(b4p),300,2))
for i,f in enumerate(b4p):
    arr1p[i] = np.loadtxt(f,delimiter =  ', ')
```

```python
plt.errorbar(arr0p[0,:,0],np.mean(arr0p[:,:,1],axis=0)/10000,yerr=np.std(arr0p[:,:,1]/10000,axis=0)**2)
```
