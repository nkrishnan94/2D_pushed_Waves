```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import glob
het_files = glob.glob("run_sims/het*.txt")
pop_files = glob.glob("run_sims/pop*.txt")
prof_files = glob.glob("run_sims/prof*.txt")
```

```python
K10000_B0 =[]
for f in het_files:
    if ("K10000_" in f) and ("B0_" in f):
        K10000_B0.append(f)
        
        
K10000_B2 =[]
for f in het_files:
    if ("K10000_" in f) and ("B2_" in f):
        K10000_B2.append(f)
        
        
        
K10000_B6 =[]
for f in het_files:
    if ("K10000_" in f) and ("B6_" in f):
        K10000_B6.append(f)
```

```python
K10000_B0_hets = np.zeros((len(K1000_B0),1000))
for i,f in enumerate(K10000_B0):
    K10000_B0_hets[i,:]= np.loadtxt(f,delimiter =  ', ')[:,1];
    
    
    
K10000_B2_hets = np.zeros((len(K10000_B2),1000))
for i,f in enumerate(K10000_B2):
    K10000_B2_hets[i,:]= np.loadtxt(f,delimiter =  ', ')[:,1];
    
    
K10000_B6_hets = np.zeros((len(K10000_B6),1000))
for i,f in enumerate(K10000_B6):
    K10000_B6_hets[i,:]= np.loadtxt(f,delimiter =  ', ')[:,1];
    
```

```python
fig,ax = plt.subplots()
t =np.arange(0,10000,10)
ax.plot(t,np.mean(K10000_B0_hets,axis=0),'navy',lw=4,label = "B = 0")
ax.plot(t,np.mean(K10000_B2_hets,axis=0),'grey',lw=4,label = "B = 2")
ax.plot(t,np.mean(K10000_B6_hets,axis=0),'orange',lw=4,label = "B = 6")


ax.errorbar(t, np.mean(K10000_B0_hets,axis=0),yerr = np.std(K10000_B0_hets,axis=0),alpha=.06,c='navy')
ax.errorbar(t, np.mean(K10000_B2_hets,axis=0),yerr = np.std(K10000_B2_hets,axis=0),alpha=.05,c='grey')
ax.errorbar(t, np.mean(K10000_B6_hets,axis=0),yerr = np.std(K10000_B6_hets,axis=0),alpha=.05,c='orange')
#ax.plot(np.mean(K10000_B2_hets,axis=0))
#ax.plot(np.mean(K10000_B6_hets,axis=0))

ax.set_ylabel('Heterozygosity',fontsize = 16)
ax.set_xlabel('Generations',fontsize = 16)
ax.legend()

for axis in ['top','bottom','left','right']:
  ax.spines[axis].set_linewidth(2)

for axis in ['top','right']:
  ax.spines[axis].set_linewidth(0)
```
